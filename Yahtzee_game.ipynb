```python
import random

def roll_dice(num_dice):
    return [random.randint(1, 6) for _ in range(num_dice)]

def display_dice(dice):
    print("Dice:", dice)

def reroll_dice(dice, indices):
    for index in indices:
        dice[index] = random.randint(1, 6)
    return dice

def score_dice(dice, category):
    if category == "Ones":
        return dice.count(1) * 1
    elif category == "Twos":
        return dice.count(2) * 2
    elif category == "Threes":
        return dice.count(3) * 3
    elif category == "Fours":
        return dice.count(4) * 4
    elif category == "Fives":
        return dice.count(5) * 5
    elif category == "Sixes":
        return dice.count(6) * 6
    elif category == "Three of a Kind":
        if any(dice.count(i) >= 3 for i in dice):
            return sum(dice)
        else:
            return 0
    elif category == "Four of a Kind":
        if any(dice.count(i) >= 4 for i in dice):
            return sum(dice)
        else:
            return 0
    elif category == "Full House":
        if any(dice.count(i) == 3 for i in dice) and any(dice.count(i) == 2 for i in dice):
            return 25
        else:
            return 0
    elif category == "Small Straight":
        if sorted(set(dice)) in ([1, 2, 3, 4], [2, 3, 4, 5], [3, 4, 5, 6]):
            return 30
        else:
            return 0
    elif category == "Large Straight":
        if sorted(set(dice)) in ([1, 2, 3, 4, 5], [2, 3, 4, 5, 6]):
            return 40
        else:
            return 0
    elif category == "Yahtzee":
        if dice.count(dice[0]) == 5:
            return 50
        else:
            return 0
    elif category == "Chance":
        return sum(dice)

def display_available_categories(categories):
    print("Available categories for scoring:")
    for category in categories:
        print("-", category)

def main():
    print("Welcome to Yahtzee!")
    num_rounds = 13
    total_score = 0
    categories = ["Ones", "Twos", "Threes", "Fours", "Fives", "Sixes",
                  "Three of a Kind", "Four of a Kind", "Full House",
                  "Small Straight", "Large Straight", "Yahtzee", "Chance"]

    for round in range(1, num_rounds + 1):
        print(f"Round {round}:")
        dice = roll_dice(5)
        display_dice(dice)

        reroll = input("Enter dice indices to reroll (space separated), or 'none' to keep all: ")
        if reroll.lower() != 'none':
            indices = [int(i) - 1 for i in reroll.split()]
            dice = reroll_dice(dice, indices)
            display_dice(dice)

        display_available_categories(categories)
        category = input("Choose a category to score: ")
        round_score = score_dice(dice, category)
        print("Round score:", round_score)
        total_score += round_score

        categories.remove(category)

    print("Game over! Total score:", total_score)

if __name__ == "__main__":
    main()
```

**These are some testing strategies:**

Random Testing: Generate random inputs and verify if the outputs are within expected ranges.

Boundary Testing: Test the game with minimum and maximum possible inputs to ensure it handles edge cases correctly.

Equivalence Partitioning: Group inputs into equivalence classes and ensure that the program behaves the same for any input within the same class.

State Testing: Verify if the game state transitions correctly between different stages (e.g., rolling dice, selecting dice to reroll, scoring).

Error Handling Testing: Intentionally provide incorrect or invalid inputs and ensure that the program handles them gracefully without crashing.

Performance Testing: Test the game with a large number of rounds or players to ensure it performs well and doesn't slow down or crash under load.
